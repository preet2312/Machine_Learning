# 'Logistic Regression' on Wine Dataset from sklearn 

### Import all the libraries and 'Wine' dataset from sklearn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

from sklearn.datasets import load_wine
wine_data = load_wine()
```

### Check the type of Wine Dataset

```python
type(wine_data)
```

### Check the keys in the dataset

```python
wine_data.keys()
```

### Check the Independent variables

```python
wine_data.feature_names
```

### Find the name of the dependent variable

```python
wine_data.target_names
```

### Check the values of the independent variables

```python
wine_data.data[:2]
```

### Values of the dependent variables

```python
wine_data.target
```

### Create a dataset with all independent values for better understanding

```python
wine = pd.DataFrame(wine_data.data,columns=wine_data.feature_names)
wine.head()
```

### Add target column in the dataset & Print the final dataset.

```python
wine['target'] = wine_data.target
```

```python
wine.head()
```

### Check for any null values in our data

```python
wine.isnull().sum()
```

### As we can see, we don't have any null values in our data

### Assign all independent values to 'x' & dependent value to 'y'

```python
x = wine_data.data
y = wine_data.target
```

### Scatter Plot to understand the relation between dependent variable and independent variable

```python
# Plot scatterplot of all the columns to target column

col = wine_data.feature_names

count=1
plt.subplots(figsize=(30, 30))
y1 = wine[['target']]

#print(y1)
for i in range(0,len(col)):
    x1 = wine[[col[i]]]
    plt.subplot(5,3,count)
    plt.scatter(x1,y1)
    plt.xlabel([col[i]])
    #print(x1)
    count+=1
```

## Check for the Correlation between all the columns.
### Find the most affected columns to target column

```python
plt.figure(figsize=(15,10))
correlation_matrix = wine.corr().round(2)
#correlation_matrix
sns.heatmap(data=correlation_matrix, annot=True)
```

### We can see that some features are positively correlated and some are negatively correlated

### We will consider all the feature to train our data, and predict the target value based on all features.

### Now split the dataset into 'Train' & 'Test' set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.25)
```

### Check the shape of the training and testing sets

```python
print(X_train.shape)
print(X_test.shape)
```

### Import Logistic Regression from sklearn

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

### Traing the model with our training dataset

```python
model.fit(X_train,Y_train)
```

### After training the model, let's predict the value based on the test set

```python
Y_predict = model.predict(X_test)

print(Y_predict)
```

### Compare above predicted value with our actual test set

```python
print(Y_test)
```

### Check the score of our trained model

```python
model.score(X_test,Y_test)
```

```python
labels = list(wine_data.target_names)
```

### Check classification report by importing it from sklearn library

```python
from sklearn.metrics import classification_report
cr = classification_report(Y_test,Y_predict,target_names=labels)

print("Classification Report : \n\n",cr)
```

### We can see that, we are getting every matrix evaluation values almost same..

### Precision, Recall, f1-score are very much similar..  Accuracy is very much high and reliable.

### Import confusion matrix from sklearn library and print the matrix comparing the actual and predicted values

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,Y_predict)

print("Confusion Matrix : \n\n",cm)
```

### Create a visualization of our confusion matrix

```python
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot()

mv = ax.matshow(cm)
plt.title('Confusion matrix Visualization')
fig.colorbar(mv)
ax.set_xticklabels([''] + labels)
ax.set_yticklabels([''] + labels)
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.show()
```

### Check our matrix accuracy

```python
from sklearn.metrics import accuracy_score
ascore = accuracy_score(Y_test,Y_predict)

print("Accuracy Score : \n\n",ascore)
```

### We can say that, our logistic model is very much accurate.. And we can predict values very efficiently based on that
