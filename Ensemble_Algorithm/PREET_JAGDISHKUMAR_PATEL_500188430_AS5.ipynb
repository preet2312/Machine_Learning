## We will take 'Pclass', 'Age' & 'Sex_male' as our dependent columns
### As they are highly correlated with our target column

```python
X = pd.DataFrame([df['Pclass'],df['Age'],df['Sex_male']]).transpose().to_numpy()
Y = df['Survived'].to_numpy()

print(X[:5])
print(Y[:5])
```

### As from the data above, we can see that all columns are not in one scale.. and model will find difficulty to understand the scale difference..

### Mean of all columns have very much difference.
### Standard deviation is also not close.

```python
X.mean(axis=0)
```

```python
X.std(axis=0)
```

### 
### We will use StandardScaler to scaler our values

```python
ss = StandardScaler()
```

```python
X = ss.fit_transform(X)
X
```

### Now, we can see the mean and standard deviation is very close to each other

```python
X.mean(axis=0)
```

```python
X.std(axis=0)
```

## 
## Split the dataset in training and testing sets

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2)
```

#### Check shape of training and testing sets

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

## Train Random Forest algorithm

```python
random_forest = RandomForestClassifier(n_estimators = 10, random_state = 0)
random_forest.fit(X_train,Y_train)

Y_predict = random_forest.predict(X_test)
```

```python
random_forest2 = RandomForestClassifier(n_estimators = 100, random_state = 0)
random_forest2.fit(X_train,Y_train)

Y_predict2 = random_forest2.predict(X_test)
```

```python
random_forest3 = RandomForestClassifier(n_estimators = 250, random_state = 0)
random_forest3.fit(X_train,Y_train)

Y_predict3 = random_forest3.predict(X_test)
```

### Compare the predicted values 

```python
dt = {'Values':['Value1','Value2','Value3','Value4','Value5','Value6','Value7'],
      'Actual':Y_test[:7],
     'RF 10 Estimators':Y_predict[:7],
     'RF 100 Estimators':Y_predict2[:7],
     'RF 250 Estimators':Y_predict3[:7]}
pred_test = pd.DataFrame(dt)
pred_test
```

## Errors Comparison

```python
matrix_evaluation = pd.DataFrame({'Kernal' :['10 Estimators','100 Estimators','250 Estimators'],
                                'MAE': [mean_absolute_error(Y_test,Y_predict),
                                       mean_absolute_error(Y_test,Y_predict2),
                                       mean_absolute_error(Y_test,Y_predict3)], 
                                'MSE': [mean_squared_error(Y_test,Y_predict),
                                       mean_squared_error(Y_test,Y_predict2),
                                       mean_squared_error(Y_test,Y_predict3)],
                                'RMSE': [np.sqrt(mean_squared_error(Y_test,Y_predict)),
                                        np.sqrt(mean_squared_error(Y_test,Y_predict2)),
                                        np.sqrt(mean_squared_error(Y_test,Y_predict3))],
                                'R2_score': [r2_score(Y_test,Y_predict),
                                            r2_score(Y_test,Y_predict2),
                                            r2_score(Y_test,Y_predict3)],
                                'Accuracy Score': [accuracy_score(Y_test,Y_predict),
                                                  accuracy_score(Y_test,Y_predict2),
                                                  accuracy_score(Y_test,Y_predict3)] })


matrix_evaluation
```

### We can clearly see that, it gives good accuracy when the estimators are more.
### MAE, MSE are very less when number of estimators are 250.
### After certain number of estimators, algorithm will give similar accuracy
### 

## Confusion matrix

```python
cml = confusion_matrix(Y_test, Y_predict)
cml2 = confusion_matrix(Y_test, Y_predict2)
cml3 = confusion_matrix(Y_test, Y_predict3)

print("\nConfusion matrix of Random Forest:\n",cml)
print("\nConfusion matrix of Random Forest:\n",cml2)
print("\nConfusion matrix of Random Forest:\n",cml3)
```

## Graphical representation of Confusion matrix

```python
labels = ['Survived','Dead']

cm = [cml,cml2,cml3]

kname = ['10 Estimators','100 Estimators','250 Estimators']


fig = plt.figure(figsize=(25,20))

for i in range(0,3):
    axes = fig.add_subplot(3,1,i+1)
    mv = axes.matshow(cm[i])
    axes.set_title('Confusion matrix Visualization '+kname[i],size='16')
    fig.colorbar(mv)
    #axes.set_xticklabels([''] + labels,size='12')
    #axes.set_yticklabels([''] + labels,size='12')
    axes.set_xlabel('Predicted Values',size='14')
    axes.set_ylabel('Actual Values',size='14')
```

## 
## Classification Report

```python
crl = classification_report(Y_test,Y_predict,target_names=labels)
crl2 = classification_report(Y_test,Y_predict2,target_names=labels)
crl3 = classification_report(Y_test,Y_predict3,target_names=labels)

print("\n\nClassification Report of 10 Estimators Random Forest: \n\n",crl)
print("\n\nClassification Report of 100 Estimators Random Forest: \n\n",crl2)
print("\n\nClassification Report of 250 Estimators Random Forest: \n\n",crl3)
```

### More accuracy and f1 score for 250 estimators.
### Not much difference between 100 and 250.
###
